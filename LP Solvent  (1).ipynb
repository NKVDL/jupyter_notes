## Let's automate labeling of withdrawal requests from LP-token holders 

```python
# Importing the libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import time
import random
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

### Let's simulate a dataset 

```python
# Simulate dataset 
center_box = (100, 200) # defines the box that cluster centres are allowed to be in
dataset, classes = make_blobs(n_samples=110000, 
                              n_features=2, 
                              centers=4,
                              center_box = (100000, 30000), # defines the box that cluster centres are allowed to be in
                              cluster_std=3424, 
                              random_state=5)

# Convert to df
df = pd.DataFrame(dataset, columns=['$TOKEN', 'TIME_AllOCATION_IN_MS']) #LP-token; amount per t time, time allocated in custody pool
df.head(3)
```

We now have dataset with users indexed $0$ to $n$ with a random token amount and random time of token allocation in the LP-pool. Let's visualise how many kernels we should use given the distribution in our simulated dataset. We use PCA to reduce the data dimensionality into the smallest variance components that explain variance in our token occurences per $j$ user per $t$ time. 

```python
# Visualise dimensionality reduction
from yellowbrick.cluster import KElbowVisualizer
model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,12)).fit(df)
visualizer.show()
```

We find an optimal of 3 latent variables that explain most of the variance in the observed tokens in the liquidity pool. So, we will initialise 3 means, or map 3 clusters where our data hangs under. Now let's create a function to calculate our latency error, and runs a k-means clustering algorithm if the NR < 2% of the total requests at time $t$. 

```python
def latency_error(amount_queried_tokens, pool_tot_tokens, queried_requests, tot_lp_token_holders, dataframe, n_clusters, t0_requests):
    
    # fraction of token request against pool
    frac_tok_req = (amount_queried_tokens / pool_tot_tokens)*100 

    # fraction of token holder requests against pool
    frac_n_req = (queried_requests / tot_lp_token_holders)*100 # How many are estimated to leave at time t

    # fraction of token amount against total token supply divided over fraction of total request against the total n of user j (per t time)
    limit = frac_tok_req / frac_n_req 

    # simple latency error (v1)
    latency_error = (t0_requests / q0_n)*100 

    
    if latency_error <= 2:
        print(f" Fraction of token request against pool, {frac_tok_req} ","\n\n"
              f" Fraction of token holder requests against pool, {frac_n_req}","\n\n"
             f" Fraction of token amount against total token supply divided over fraction of total request against the total n of user j (per t time) {limit}","\n\n"
             f" Latency error, {latency_error}","\n\n"
             f"Initiating k-means...")
        
        #### CLUSTERING ####
        start = time.time()
        # Cluster data a give t time instance
        kmeans = KMeans(n_clusters=n_clusters, init='k-means++', random_state=0).fit(df)
        end = time.time()
        total_time_ms = end - start
        
        print(f"\nThe total time to cluster LP-token holders into color-coded categories to limit their withdrawal: {total_time_ms} milliseconds")
        
        # Show assigned labels
        sns.scatterplot(data=dataframe, x="$TOKEN", y="TIME_AllOCATION_IN_MS", hue=kmeans.labels_)
        plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], 
            marker="X", c="r", s=80, label="centroids")
        plt.legend()
        plt.show()
        
        # Get total within sum variance for ni state of the network - one n
        print(f"This is the total within sum variance for ni state of the network, {kmeans.inertia_}")
        
        # Get label per j user with yi tokens and time of holding token  per n state of the network
        cluster_labels = kmeans.labels_

        # Add labels to user j with yi tokens
        dataframe['Color-code'] = cluster_labels

        #### CLUSTERING ####

        token_values = dataframe['$TOKEN'].values 

        X = dataframe.iloc[:,[0,1]].values
        
        return dataframe

        

    else:
        print(f"The latency error > 2%: {latency_error}%%")
    

```

Now that we have our function let's simulate LP token holders, total tokens in LP-pool, new requests at time t, newqueried_requests, total amount of queried requests.

```python
t0_n = random.randint(0, 110000) #total amount of custody pool allocators
t0_y = random.randint(0, 1000000) #total amount of tokens in  custody pool 
t0_req = random.randint(0, 100) #new requests at time t
q0_n = random.randint(0, 7000) # queried requests
q0_t = random.randint(0, 4000) #amount of LP-tokens queried 
```

```python
latency_error(q0_t, t0_y, q0_n, t0_n, df, 3, t0_req) 
```
